```python
import os

import pandas as pd
```

```python
titanic_data = pd.read_csv(os.path.join('titanic.csv'))
```

```python
titanic_data.info()
```

```python
titanic_data.head()
```

```python
titanic_data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace=True)
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
titanic_data['Sex'] = le.fit_transform(titanic_data['Sex'].astype(str))
titanic_data.head()
```

```python
titanic_data = pd.get_dummies(titanic_data, columns=['Embarked'])
titanic_data.head()
```

```python
titanic_data[titanic_data.isnull().any(axis=1)]
```

```python
titanic_data = titanic_data.dropna()
```

```python
from sklearn.cluster import MeanShift

analyzer = MeanShift(bandwidth=30)
analyzer.fit(titanic_data)
```

```python
from sklearn.cluster import estimate_bandwidth
estimate_bandwidth(titanic_data)
```

```python
labels = analyzer.labels_
labels
```

```python
import numpy as np
np.unique(labels)
```

```python
titanic_data['cluster_group'] = labels
titanic_data
```

```python
titanic_data.describe()
```

```python
titanic_data.groupby(['cluster_group']).mean()
```

```python
titanic_df = titanic_data.groupby(['cluster_group']).mean()
titanic_df
```

```python
titanic_df['Counts'] = titanic_data.groupby(['cluster_group'])['Sex'].count()
```
