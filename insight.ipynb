```python
import os
import pandas as pd
```

```python
scoring = pd.read_pickle(os.path.join('dataset01', 'scoring_mi.pickle'))
```

```python
scoring.head()
```

```python
wayne = scoring.loc['gretzwa01']
wayne
```

```python
sums = wayne.groupby(level=0)[['G','A']].sum()
```

```python
sums
```

```python
wayne.index
```

```python
wayne.index.nlevels
```

```python
means = sums.rolling(3)['G'].mean()
```

```python
means
```

```python
%matplotlib inline
sums.assign(roll_mean = means).plot()
```

```python
means = sums.rolling(3, min_periods=2)['G'].mean()
means.head(10)
```

```python
means = sums.rolling(3).mean()
means
```

```python
means = sums.rolling(3, center=True).mean()
means
```
