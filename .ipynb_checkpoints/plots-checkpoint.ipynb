```python
df = pd.read_csv(os.path.join('dataset01', 'sales.csv'), parse_dates=['Date'])
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df['Month'] = df['Date'].dt.month
df
```

```python
df.drop(columns='Date', inplace=True)
```

```python
df
```

```python
sales = df.groupby(['Month', 'Product'])['Amount'].sum().unstack()
sales
```

```python
sales.plot()
```

```python
styles = {'a': ':s', 'b': '--v', 'c': '-.o'}

with plt.style.context('ggplot'):
    fig, ax = plt.subplots(figsize=(6, 3))
    for column, style in styles.items():
        sales[column].dropna().plot(ax=ax, style=style, grid=True, legend=True, label=f'Product {column}')
        
ax.set_title('Monthly sales')
ax.set_ylabel('Sales (EUR)')

ax.set_xticks(range(1, 13))

import calendar
ax.set_xticklabels([calendar.month_name[x] for x in sales.index], rotation=30, ha='right')

from matplotlib.ticker import FuncFormatter

ax.yaxis.set_major_formatter(FuncFormatter(lambda tick_value, pos: f'{tick_value / 1000}k'))

fig.tight_layout() # wyrównanie jeśli coś na siebie zachddzi lub się źle wyświatla
```

```python
df = pd.read_csv(os.path.join('dataset01', 'athletes.csv'), parse_dates=['date_of_birth'])
df.head()
```

```python
df.shape
```

```python
f = df[df['sex'] == 'female']
m = df[df['sex'] == 'male']

tennis_f = f[f['sport'] == 'tennis']
tennis_m = m[m['sport'] == 'tennis']

gym_f = f[f['sport'] == 'gymnastics']
gym_m = m[m['sport'] == 'gymnastics']
```

```python
fig, ax = plt.subplots()

plots = [
    (tennis_f, 'red', 'Tennis women'),
    (tennis_m, 'blue', 'Tennis men'),
    (gym_f, 'orange', 'Gymnastics women'),
    (gym_m, 'green', 'Gymnastics men'),
]

for df, color, label in plots:
    df.plot.scatter(ax=ax, x='weight', y='height', alpha=.25, color=color, label=label)
    
ax.set_xlabel('Weight (kg)')
ax.set_ylabel('Height (m)')
ax.set_title('Height and weight of athletes')

fig.tight_layout()

```

```python
fig, ax = plt.subplots()
gym_m.plot.scatter(ax=ax, x='height', y='weight', c='gold', colormap='plasma')
fig.tight_layout()
```

```python
df = pd.read_csv(os.path.join('dataset01', 'sales.csv'), parse_dates=['Date'])
```

```python
df.head()
```

```python
df['Month'] = df['Date'].dt.month
```

```python
df.drop(columns='Date', inplace=True)
```

```python
g = df.groupby(['Month', 'Product'])['Amount'].sum().unstack()
g
```

```python
revenue = g.fillna(0)
revenue
```

```python
revenue = revenue.cumsum()
revenue
```

```python
fig, ax = plt.subplots(figsize=(12, 9))


total_revenue = df.groupby('Month')['Amount'].sum().cumsum()
total_revenue.plot(ax=ax, style='r--', legend=True, label='All')
revenue.plot.area(title='Revenue 2024 [EUR]', ax=ax, stacked=False, grid=True)


fig.tight_layout()
```

```python
df = pd.read_csv(os.path.join('dataset01', 'weather.csv'))
df.head()
```

```python
df.shape
```

```python
df
```

```python
days = df[(df['MONTH'].isin([1, 7])) & (df['DAY'] == 1)].drop(columns='DAY')
days
```

```python
days = days.pivot(columns='MONTH', index='TIME')
days.head()
```

```python
fig, ax = plt.subplots()

days['TEMP'].plot(ax=ax)
days['PRESSURE'].plot(ax=ax)
```

```python
fig, ax = plt.subplots(2, 2)
```

```python
plt.style.use('ggplot')
```

```python
fig, ax = plt.subplots(2, 2, sharey='row', sharex='col')

days['TEMP'].plot(ax=ax[0], subplots=True, legend=False)
days['PRESSURE'].plot(ax=ax[1], subplots=True, legend=False)

ax[0][0].set_ylabel('Temperature')
ax[0][0].set_title('January')
ax[0][1].set_title('July')
ax[1][0].set_xlabel('Time')
ax[1][1].set_xlabel('Time')

ax[1][0].set_ylabel('Pressure')
```

```python
df = pd.read_csv(os.path.join('dataset01', 'athletes.csv'))
```

```python
def make_plot(sport1, gender1, sport2, gender2):
    fig, ax = plt.subplots()
    a = df[(df['sex'] == gender1) & (df['sport'] == sport1)]
    b = df[(df['sex'] == gender2) & (df['sport'] == sport2)]
    
    a.plot.scatter(ax=ax, x='weight', y='height', alpha=0.25, color='b', label=f'{sport1} {gender1}')
    b.plot.scatter(ax=ax, x='weight', y='height', alpha=0.25, color='r', label=f'{sport2} {gender2}')  
    
    
    
```

```python
all_sports = df['sport'].drop_duplicates().sort_values()
```

```python
interact(make_plot, sport1=all_sports, gender1=['male', 'female'], sport2=all_sports, gender2=['male', 'female'])
```

```python
df = pd.read_csv(os.path.join('dataset01', 'athletes.csv'))
medals = df.groupby(['sport', 'nationality'])[['gold', 'silver', 'bronze']].sum()
medals.sort_values(['gold'], ascending=False, inplace=True)
medals
```

```python
def make_plot(n=10, sport='aquatics'):
    fig, ax = plt.subplots(figsize=(n, 6))
    top = medals.loc[sport].head(n)
    
    top.plot.bar(ax=ax, color=['gold', 'silver', 'brown'], stacked=True, rot=45, title=f'Top {n} countries in {sport}')
    ax.set_ylabel('medals')
```

```python
interact(make_plot, n=widgets.IntSlider(min=5, max=100, continuous_update=False), sport=all_sports)
```

```python
def x(ok=True):
    return 'ok' if ok else 'no'
```

```python
interact(x, ok=widgets.Checkbox(value=False))
```
