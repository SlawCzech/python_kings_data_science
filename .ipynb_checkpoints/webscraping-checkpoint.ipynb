```python
def convert_html_into_dict(arr, driver):
    articles = []
    
    for article in arr:
        driver.execute_script('arguments[0].scrollIntoView()', article)
        driver.implicitly_wait(1)
        title = article.find_element(By.CSS_SELECTOR, 'h2').text
        link = article.find_element(By.CSS_SELECTOR, 'a').get_attribute('href')
        img_url = article.find_element(By.CSS_SELECTOR, 'img').get_attribute('src')
        
        if not img_url:
            img_url = article.find_element(By.CSS_SELECTOR, 'img').get_attribute('srcset')
        
        articles.append({'title': title, 'link': link, 'img_url': img_url})
    
    return articles
```

```python
driver = webdriver.Chrome(executable_path='driver/chromedriver')
driver.get('http://tvn24.pl')

driver.implicitly_wait(5)
css = By.CSS_SELECTOR

rodo = driver.find_element(By.CSS_SELECTOR, '#onetrust-accept-btn-handler')
rodo.click()

articles_html = driver.find_elements(css, '.teaser-wrapper > article:not(.html-teaser--standard, .gallery-teaser--standard ,.external-teaser--standard)')


results = convert_html_into_dict(articles_html, driver)
print(results)

# print(articles_html) 
# len(articles_html)


driver.close()
```
